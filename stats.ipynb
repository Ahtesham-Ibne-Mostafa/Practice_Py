```python
# problem 5.1

def main(x):
    
    x=x[1:-1]
    x=x.split(',')
    lyst=[]
    for number in x:
        lyst.append(int(number))

    print('List: ',lyst)
    mode(lyst)
    median(lyst)
    mean(lyst)
def mode(lyst):
    x=lyst.copy()
    a=0
    b=''
    for i in x:
        f=x.count(i)
        if f>a:
            a=f
            b=str(i)
            x.remove(i)
        elif f==a:
            b+=', '+str(i)
    print('Mode: ',b)

def median(lyst):
    notun=sorted(lyst)
    if len(notun)%2==0:
        sum_of_middle_two_number=notun[len(notun)//2-1]+notun[len(notun)//2]
        moddhok=sum_of_middle_two_number/2
    else:
        moddhok=notun[len(notun)//2+1]
    print('Median: ',moddhok)
    
def mean(lyst):
    somosti=sum(lyst)
    gor=somosti/len(lyst)
    print('Mean: ',gor)
    
x=input('enter the list :')
if x[1]!=']':
    test=main(x)
else:
    print(0)
```

```python
# problem 5.2

f = open("demofile.txt", "r")
x=[]
for i in f:
    p=i.split('\n')
    x.append(p[0])
z=len(x)
if z>1:
    print('This file has',z,'lines.')
else:
    print('This file has',z,'line.')
y=1
while y!=0:
    y=int(input("Enter a line number [0 to quit]: "))
    if y<=z and y!=0:
        print(y,':',x[y-1])
    elif y>0:
        print('ERROR: line number must be less than',z)

f.close()
```

```python
# Problem 5.7

f = open("demofile.txt", "r")
lst=[]
for i in f:
    a=''
    for j in i:
        if j ==' ' or j=='\n':
            if a not in lst:
                lst.append(a)
            a=''
            
        else:
            a+=j
if a not in lst:
    lst.append(a)
    
new=sorted(lst)
for i in new:
    print(i)
```

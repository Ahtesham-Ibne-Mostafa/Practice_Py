```python
string=input('Enter the lines :')
vawel='aeiouAEIOU'
num1='0123456789'
ky=0
v=0
c=0
count=0
list1=[]
dic={}
for i in string :
    count+=1
    if i==',' or count==len(string):
        if count==len(string):
            if i in vawel:
                v+=1
            else:
                c+=1
        ky+=1
        list1.append(v)
        list1.append(c)
        dic[ky]=tuple(list1)
        list1=[]
        v=0
        c=0
    elif i in vawel :
        v+=1
    elif i not in vawel and i!=' ' and i not in num1:
        c+=1
print(dic)
```

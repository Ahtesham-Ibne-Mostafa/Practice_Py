```python
x= int(input('enter the number :'))
for y in range (x):
    if y==0 :
        print('*'*x)
    elif y==x-1 :
        print('*'*x)
    else :
        print(' '*(x-y-1)+'*') 
```

```python
x=int(input('enter the number'))
for i in range(1,x+1):
    if i%2==0:
        print('\n== Iteration',i,'==')
        for j in range(1,i+1):
            print('x'*j)
    else :
        print('\n== Iteration',i,'==')
        for k in range (1,i+1):
            for k in range (1,i+1):
                print(k,end='')               
            print('\n')
```

```python
n=int(input('enter the number :'))
m=n-1
o=1
for x in range (1,n*2):
    if x>=n+1:
        print(' '*m+'*'*o )
        m+=1
        o-=2
    else :
        print(' '*m+'*'*o )
        m-=1
        o+=2
        if x==n:
            o-=4
            m+=2
```

```python
h = int(input())
for i in range(h):
    print(" "*((h-i)-1)+"*"*(i*2+1))
for i in reversed(range(h-1)):
    print(" "*((h-i)-1)+"*"*(i*2+1)) 
```
